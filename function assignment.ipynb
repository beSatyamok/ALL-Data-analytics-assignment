1.What is difference between a function and a method  in python?

Ans:Function in Python

. A function is a block of reusable code that performs a specific task.

. It is defined using the def keyword.

.It can exist independently (outside of any class).

. Example of a function:

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
```

. Method in Python:

. A method is also a function, but it is associated with an object (it belongs to a class).

.It is defined inside a class and automatically takes the instance (self) as its first argument.

.You call a method using the dot (.) notation on an object.

. Example of a method:

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):  # This is a method
        return f"Hello, my name is {self.name}."

p = Person("Alice")
print(p.greet())
```

| Feature        | Function                            | Method                                     |
| -------------- | ----------------------------------- | ------------------------------------------ |
| Belongs to     | Independent (not tied to an object) | Associated with a class/object             |
| Defined with   | `def` keyword                       | `def` keyword inside a class               |
| First argument | Normal arguments only               | Always takes `self` (for instance methods) |
| Call style     | `function_name(args)`               | `object.method(args)`                      |

5.What are iterators in python and how do they differ from  iterables?

Ans.1. What is an Iterable:
. An iterable is any Python object capable of returning its members one at a time.

. Examples: list, tuple, string, dictionary, set.

. Technically, an iterable is any object that implements the __iter__() method (which returns an iterator).

. Example:

```python
numbers = [1, 2, 3, 4]
for num in numbers:   # list is iterable
    print(num)
```

2. What is an Iterator?

.An iterator is an object used to iterate (traverse) through an iterable.

.It keeps track of where you are in the iteration.

.It implements two methods:

__iter__() → returns the iterator object itself.

__next__() → returns the next item in the sequence. Raises StopIteration when items are exhausted.

. Example:

```python
numbers = [1, 2, 3]
it = iter(numbers)   # get iterator from iterable

print(next(it))  # 1
print(next(it))  # 2
print(next(it))  # 3
# print(next(it))  # Raises StopIteration
```

| **Aspect**      | **Iterable**                                 | **Iterator**                                        |
| --------------- | -------------------------------------------- | --------------------------------------------------- |
| **Definition**  | Object that can return an iterator.          | Object used to iterate through iterable.            |
| **Methods**     | Must implement `__iter__()`                  | Must implement both `__iter__()` and `__next__()`   |
| **Examples**    | List, tuple, set, dict, string               | Objects returned by `iter()`, e.g., `list_iterator` |
| **Reusability** | Can be reused (you can call `iter()` again). | Cannot be reused once exhausted.                    |

6.Explain the concept of gernators in python and how they are defined.

Ans.A generator in Python is a special type of iterator that is used to produce a sequence of values lazily (on-demand), instead of generating and storing all the values in memory at once.

This makes generators memory efficient and useful for working with large data streams or infinite sequences.

How are Generators Defined?

. Generators can be created in two main ways:

1. Using a Function with yield

. A normal Python function uses return to send a result back.

. A generator function uses yield instead of return.

. Each time the function is called, execution is paused at yield and can be resumed later.

. Example:

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count   # pauses here and returns value
        count += 1

# Using the generator
for number in count_up_to(5):
    print(number)
```

2. Using Generator Expressions

These are similar to list comprehensions, but they use parentheses () instead of square brackets [].

Example:

```python
squares = (x * x for x in range(5))
print(next(squares))  # 0
print(next(squares))  # 1
print(next(squares))  # 4
```

7.What  is the  advantages of  using gernators over regular function.

Ans:  Regular Function

A normal function with return will compute all results at once and usually store them in a list (or other structure) before returning.

. Example:

```python
def get_squares(n):
    result = []
    for i in range(n):
        result.append(i * i)
    return result

print(get_squares(5))
```

. Generator Function:

.A generator with yield will produce values one at a time and does not keep the entire list in memory.

. Example:

```python
def get_squares_gen(n):
    for i in range(n):
        yield i * i

for val in get_squares_gen(5):
    print(val)
```

. Advantages of Generators over Regular Functions

1.Memory Efficiency

. Generators do not store the entire sequence in memory.

. Useful when dealing with large datasets or infinite streams.

. Example: reading a huge file line by line instead of loading it entirely.

2.Lazy Evaluation (On-Demand Computation)

. Values are produced only when needed.

. Saves both CPU time and memory if you don’t need all results at once.

3.Infinite Sequences Possible

. A regular function cannot return an "infinite list", but a generator can yield values endlessly.

.Example:

```python
def infinite_numbers():
    num = 1
    while True:
        yield num
        num += 1
```

4.Cleaner Code

. Generators allow writing simple loops instead of building and returning large lists.

. Avoids boilerplate like managing temporary result lists.

5.Better Performance for Iteration

. For large loops, generators avoid the overhead of creating and storing all elements at once.

9.Explain the purpose and usage of the map() function in python.

Ans. The map() function in Python is used to apply a given function to each item of an iterable (like a list, tuple, or string) and returns an iterator (which can be converted to a list, set, or tuple).

. Purpose of map()

.To perform an operation on every element of an iterable without writing explicit loops.

.It makes code cleaner and more functional programming–oriented.

.Useful when you need to transform data (e.g., converting strings to integers, squaring numbers, etc.).

. function → the function to apply to each element.

. iterable → one or more iterables (map can accept multiple).

. Returns a map object (an iterator).

. Basic Example

```python
# Square each number in a list
numbers = [1, 2, 3, 4, 5]

def square(n):
    return n * n

result = map(square, numbers)
print(list(result))
```

. Using map() with lambda

```python
numbers = [10, 20, 30, 40]
result = map(lambda x: x / 10, numbers)
print(list(result))
```

. Using map() with multiple iterables

.If multiple iterables are given, map() applies the function to elements pairwise:



```python
a = [1, 2, 3]
b = [4, 5, 6]

result = map(lambda x, y: x + y, a, b)
print(list(result))
```

5.Create an iterator class in python that gernate the fibbonacci sequence upto a specified numbers of terms.

```python
class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # total number of terms to generate
        self.count = 0          # counter to track how many terms have been generated
        self.a = 0              # first Fibonacci number
        self.b = 1              # second Fibonacci number

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            fib = self.a + self.b
            self.a, self.b = self.b, fib
            self.count += 1
            return fib

# Example usage
n = 10  # generate first 10 Fibonacci numbers
fib_sequence = Fibonacci(n)
for num in fib_sequence:
    print(num, end=" ")
```

```python
def read_file_line_by_line(file_path):
    """
    Generator function that reads a file line by line
    and yields each line as a string.
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Remove newline character
    except FileNotFoundError:
        print(f"The file {file_path} does not exist.")

# Example usage:
file_path = 'example.txt'

for line in read_file_line_by_line(file_path):
    print(line)
```
